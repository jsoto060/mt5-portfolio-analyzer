# 02 What-If Analysis
Thin notebook: scenario orchestration using ReplayAnalyzer facade.

```python
import os
import sys
from IPython.display import display

cwd = os.getcwd()
candidates = [cwd, os.path.dirname(cwd)]
REPO = next((p for p in candidates if os.path.exists(os.path.join(p, 'src'))), cwd)
SRC = os.path.join(REPO, 'src')
if SRC not in sys.path:
    sys.path.insert(0, SRC)

from replay_analyzer import default_analyzer_for_repo
import comparison
```

```python
analyzer = default_analyzer_for_repo(REPO)
folder = os.path.join(REPO, 'data', 'baseline')

baseline = analyzer.replay_folder(folder)

scenario = {
    'GBPUSD': {
        'risk_percent': 1.8,
        'max_trades': 3
    }
}

what_if = analyzer.replay_scenario(folder, overrides=scenario)
tables = comparison.compare_results(baseline, what_if)

print('Baseline Configuration')
display(baseline.configuration())

print('Baseline vs Scenario: Performance')
display(tables['performance'])

print('Baseline vs Scenario: Drawdown')
display(tables['drawdown'])

print('Baseline vs Scenario: Margin')
display(tables['margin'])

print('Baseline vs Scenario: Per Pair')
display(tables['per_pair'])
```

```python
baseline.plot_equity().show()
what_if.plot_equity().show()
baseline.plot_margin().show()
what_if.plot_margin().show()
baseline.plot_pair_contributions().show()
what_if.plot_pair_contributions().show()
baseline.plot_pair_floating().show()
what_if.plot_pair_floating().show()
baseline.plot_pair_drawdown().show()
what_if.plot_pair_drawdown().show()
```
