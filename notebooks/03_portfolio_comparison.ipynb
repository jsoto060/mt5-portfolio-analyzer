# 03 Portfolio Comparison
Thin notebook: baseline vs proposed folder comparison via ReplayAnalyzer.

```python
import os
import sys
from IPython.display import display

cwd = os.getcwd()
candidates = [cwd, os.path.dirname(cwd)]
REPO = next((p for p in candidates if os.path.exists(os.path.join(p, 'src'))), cwd)
SRC = os.path.join(REPO, 'src')
if SRC not in sys.path:
    sys.path.insert(0, SRC)

from replay_analyzer import default_analyzer_for_repo
```

```python
analyzer = default_analyzer_for_repo(REPO)
comparison_result = analyzer.compare_folders(
    baseline_folder=os.path.join(REPO, 'data', 'baseline'),
    proposed_folder=os.path.join(REPO, 'data', 'proposed')
)

tables = comparison_result['tables']

print('Configuration Comparison')
display(tables['configuration'])

print('Summary Comparison')
display(tables['performance'])

print('Drawdown Comparison')
display(tables['drawdown'])

print('Margin Comparison')
display(tables['margin'])

print('Per-Pair Comparison')
display(tables['per_pair'])
```
